```python
# Importing Libraries
import pandas as pd
import numpy as np 
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
import matplotlib.pyplot as plt
```

## Data Pre-processing

```python
# Renaming the column names
df.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','city':'City','latitude':'Latitude',
                   'provstate':'State','longitude':'Longitude','region_txt':'Region','attacktype1_txt':'Attack_Type',
                   'target1':'Target','nkill':'Killed','nwound':'Wounded','summary':'Summary','gname':'Group',
                   'targtype1_txt':'Target_type','weaptype1_txt':'Weapon_type','motive':'Motive'},inplace=True)
```

```python
# data for further processing
df=df[['Year','Month','Day','Country','State','Region','City','Latitude','Longitude','Attack_Type','Killed',
               'Wounded','Target','Summary','Group','Target_type','Weapon_type','Motive']]
df.head()
```

```python
# Checking for the null values 
df.isnull().sum()
```

```python
# Checking the value_counts

print("Country with the most attacks:",df['Country'].value_counts().idxmax())

# since first entry is 'unknown'
print("City with the most attacks:",df['City'].value_counts().index[1]) 
print("Region with the most attacks:",df['Region'].value_counts().idxmax())
print("Year with the most attacks:",df['Year'].value_counts().idxmax())
print("Month with the most attacks:",df['Month'].value_counts().idxmax())
print("Group with the most attacks:",df['Group'].value_counts().index[1])
print("Most Attack Types:",df['Attack_Type'].value_counts().idxmax())
```

```python
df['Year'].value_counts(dropna = False).sort_index()
```

## Data visualization

```python
# Terrorist Activities In Each Year
x_years = df['Year'].unique()
y_years = df['Year'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (15,10))

sns.barplot(x = x_years,y = y_years,palette = 'rocket')
plt.xticks(rotation = 50)
plt.xlabel('Attack Year')
plt.ylabel('Number of Attacks Each Year')
plt.title('Attack In Years')
plt.show()
```

```python
# Terrorist Activities By Region In Each Year
pd.crosstab(df.Year, df.Region).plot(kind='area',figsize=(15,10))
plt.title('Terrorist Activities By Region In Each Year')
plt.ylabel('No of Attacks')
plt.show()
```

```python
df['Killed'] = df['Killed'].fillna(0).astype(int)
df['Wounded'] = df['Wounded'].fillna(0).astype(int)
df['Casualities'] = df['Killed'] + df['Wounded']
```

```python
# Top 50 worst attacks
dff= df.sort_values(by='Casualities',ascending=False)[:50]
dff
```

```python
ht=dff.pivot_table(index='Country',columns='Year',values='Casualities')
ht.fillna(0,inplace=True)
ht.head()
```

```python
# Total Number of people killed in terror attack
killData = df.loc[:,'Killed']
print('Number of people killed by terror attack:', int(sum(killData.dropna()))) # drop the NaN values
```

```python
# Let's look at what types of attacks these deaths were made of.
attackData = df.loc[:,'Attack_Type']
# attackData
typeKillData = pd.concat([attackData, killData], axis=1)
```

```python
typeKillData.head()
```

```python
typeKillFormatData = typeKillData.pivot_table(columns='Attack_Type', values='Killed', aggfunc='sum')
typeKillFormatData
```

```python
# Number of Killed in Terrorist Attacks by Countries
countryData = df.loc[:,'Country']
# countyData
countryKillData = pd.concat([countryData, killData], axis=1)
```

```python
fig_size = plt.rcParams["figure.figsize"]
fig_size[0]=25
fig_size[1]=25
plt.rcParams["figure.figsize"] = fig_size
```

```python
countryKillFormatData = countryKillData.pivot_table(columns='Country', values='Killed', aggfunc='sum')
countryKillFormatData
```

```python
labels = countryKillFormatData.columns.tolist()
labels = labels[:50] #50 bar provides nice view
index = np.arange(len(labels))
transpoze = countryKillFormatData.T
values = transpoze.values.tolist()
values = values[:50]
values = [int(i[0]) for i in values] # convert float to int
colors = ['red', 'green', 'blue', 'purple', 'yellow', 'brown', 'black', 'gray', 'magenta', 'orange'] # color list for bar chart bar color 
fig, ax = plt.subplots(1, 1)
ax.yaxis.grid(True)
fig_size = plt.rcParams["figure.figsize"]
fig_size[0]=25
fig_size[1]=25
plt.rcParams["figure.figsize"] = fig_size
plt.bar(index, values, color = colors, width = 0.9)
plt.ylabel('Killed People', fontsize=30)
plt.xlabel('Countries', fontsize = 30)
plt.xticks(index, labels, fontsize=20, rotation=90)
plt.title('Number of People Killed By Countries', fontsize = 30)
# print(fig_size)
plt.show()
```
